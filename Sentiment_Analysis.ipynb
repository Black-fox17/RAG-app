```python
from transformers import TFAutoModel,AutoTokenizer
from datasets import load_dataset
import tensorflow as tf
```

```python
model = TFAutoModel.from_pretrained("bert-base-uncased")
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
data = load_dataset('SetFit/emotion')
```

```python
type(data)
```

```python
data
```

```python
def tokenize(batch):
  return tokenizer(batch['text'],padding = True,truncation = True)
```

```python
emotions_encoded = data.map(tokenize , batched = True, batch_size = None)
```

```python

emotions_encoded
```

```python
emotions_encoded.set_format('tf',
                            columns=['input_ids', 'attention_mask', 'token_type_ids', 'label'])
batch_size = 64
def order(inp):
  data = list(inp.values())
  return {"input_ids":data[1],
          "attention_mask" : data[2],
          "token_type_ids" : data[3]
          },data[0]

train_dataset = tf.data.Dataset.from_tensor_slices(emotions_encoded['train'][:])
train_dataset = train_dataset.batch(batch_size).shuffle(1000)
train_dataset = train_dataset.map(order,num_parallel_calls = tf.data.AUTOTUNE)

test_dataset = tf.data.Dataset.from_tensor_slices(emotions_encoded['test'][:])
test_dataset = test_dataset.batch(batch_size).shuffle(1000)
test_dataset = test_dataset.map(order,num_parallel_calls = tf.data.AUTOTUNE)

validation_dataset = tf.data.Dataset.from_tensor_slices(emotions_encoded['validation'][:])
validation_dataset = validation_dataset.batch(batch_size).shuffle(1000)
validation_dataset = validation_dataset.map(order,num_parallel_calls = tf.data.AUTOTUNE)
```

```python
class BertClassification(tf.keras.Model):
  def __init__(self,bert_model,num_classes):
    super().__init__()
    self.bert_model = bert_model
    self.num_classes = tf.keras.layers.Dense(num_classes,activation = 'softmax')

  def call(self,inputs):
    x = self.bert_model(inputs)[1]
    return self.num_classes(x)
```

```python
classifier = BertClassification(model,num_classes = 6)

classifier.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-5),
                   loss = tf.keras.losses.SparseCategoricalCrossentropy(),
                   metrics = ['accuracy'])
```

```python
history = classifier.fit(train_dataset,epochs = 3,verbose = 1)
```

```python
classifier.evaluate(test_dataset)
```
